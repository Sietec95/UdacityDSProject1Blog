## Data preparation
- Remove columns without variance (only one unique value).   
- Remove columns that will not be used in further analysis (e.g. urls, ids). Case by case explanation is provided in the comment below.   
- Replace text that can not be categorized with boolean if the text field is filled.   
~~- Replace dates with continuous variable (Timestamp)~~
- Replace t/f with True/False
- Create multiple columns from columns with lists as entries
- Prepare numerical columns by removing currency, percentage, etc.

```python
# Columns to be dropped
cols_to_drop = ["listing_url", "host_id", "host_url", "host_location", "host_total_listings_count", "street", "neighbourhood",
                "neighbourhood_cleansed", "city", "latitude", "longitude", "calculated_host_listings_count",
               "calendar_updated", "availability_30", "availability_60", "availability_90", "availability_365", "weekly_price",
               "monthly_price"]
# Columns with text that can not easily be categorized, replace with True if text field is filled, False if empty
cols_text = ["name", "summary", "space", "description", "neighborhood_overview", "notes", "transit", "thumbnail_url",
             "medium_url", "picture_url", "xl_picture_url", "host_name", "host_about", "host_thumbnail_url",
             "host_picture_url", "square_feet", "host_neighbourhood"]
cols_t_f = ["host_is_superhost", "host_has_profile_pic", "host_identity_verified", "is_location_exact", "instant_bookable",
           "require_guest_profile_picture", "require_guest_phone_verification"]
cols_with_list = ["host_verifications", "amenities"]
cols_dates = ["host_since", "first_review", "last_review"]
cols_remove_currency = ["price", "security_deposit", "cleaning_fee", "extra_people"]
cols_percentage = ["host_response_rate", "host_acceptance_rate"]
```

```python
def replace_text_with_bool(data, cols):
    """
    INPUT
    data - pandas dataframe
    cols - List of columns in data with strings
    OUTPUT
    data - pandas dataframe where all columns in cols are replaced by a new bool column
    
    For every column in cols, the column is replaced by a new column named "has_"col.
    The values in the new column are True if the original column had a string stored in the cell, otherwise it is False
    """
    for col in cols:
        if col in list(data.columns):
            data["has_"+col] = True
            data.loc[data[col].isna(), "has_"+col] = False
    data.drop(columns=cols, inplace=True)
    return data

def replace_t_f_with_bool(data, cols):
    """
    INPUT
    data - pandas dataframe
    cols - List of columns in data where boolean values are stored as text ("t", "f")
    OUTPUT
    data - pandas dataframe with cleaned boolean columns
    
    This function replaces strings "t", "f" with boolean True and False for each columns in the input list cols
    """
    for col in cols:
        data[col] = data[col].map(lambda x: True if x == 't' else False)
    return data

def get_bool_from_list(row, col, cell_item):
    """
    INPUT
    row - Row of the dataset to accessed with lambda function
    col - Original column name
    cell_item - unique value to be searched for
    OUTPUT
    val - Bool, True if cell_item is in col of the row, False otherwise
    
    Return Bool that is True if the input cell item is in the column of the row, False if it is not
    """
    val = False
    try:
        vals = row[col].replace('[', '').replace("'", '').replace("]", '').replace('"', '').replace('{', '').replace('}', '').split(',')
        if cell_item in vals:
            val = True
    except:
        val = False
    return val

def split_list_into_columns(data, cols):
    """
    INPUT
    data - pandas dataframe
    cols - List of columns with list entries to be split into multiple Boolean columns
    OUTPUT
    data - pandas dataframe with one new column for each unique value in the lists in cols

    Split columns where the cells contain lists into multiple columns, one for each unique value in the lists
    """
    for column_name in cols:
        # Create list of unique values
        uniques = list(data[column_name].unique())
        uniques = [x.split(",") for x in uniques]
        uniques = [item for sublist in uniques for item in sublist]
        remove = ['{', '}', '[', ']', '"', '"']
        uniques = pd.Series(uniques)
        for rem_str in remove:
            uniques = uniques.str.replace(rem_str, '')
        uniques = list(set(uniques))
        uniques = list(filter(None, uniques))

        # split into columns
        for unique in uniques:
            data[column_name + '_' + unique] = data.apply(lambda row: get_bool_from_list(row, column_name, unique),axis=1)
    data = data.drop(columns=cols)
    return data

def remove_currency_from_column(data, cols):
    """
    INPUT
    data - pandas dataframe
    cols - List of columns with with currency values where the currency sign needs to be removed
    OUTPUT
    data - pandas dataframe with currency columns as float

    Remove the currency ($) signs from columns and change the type to float
    """
    for col in cols:
        data[col] = data[col].str.replace(',', '').str.replace('$', '').astype(float)
    return data

def remove_percentage_from_column(data, cols):
    """
    INPUT
    data - pandas dataframe
    cols - List of columns where the percentage sign needs to be removed
    OUTPUT
    data - pandas dataframe with percentage columns as float

    Remove the percentage (%) signs from columns and change the type to float
    """
    for col in cols:
        data[col] = data[col].str.replace(',', '').str.replace('%', '').astype(float)
        data[col] = data[col]/100
    return data

def prepare_listing_data(data, cols_to_drop, cols_text, cols_t_f, cols_with_list, cols_remove_currency, cols_percentage):
    """
    INPUT
    data - pandas dataframe with listing data
    cols_to_drop - columns to be dropped without further assessment
    cols_text - Columns with text that can not easily be categorized, replace with True if text field is filled, False if empty
    cols_t_f - Columns with values "t" and "f" in the original data, to be replaced by True and False
    cols_with_list - Columns with lists as values, lists are seperated into multiple columns
    #cols_dates - dates
    cols_remove_currency - Columns with numerical values as currency, currency must be removed for use as numerical data
    cols_percentage - Columns with numerical values as percentages, percentage sings must be removed for use as numerical data
    OUTPUT
    data - pandas dataframe with all preparation steps carried out
    
    Carry out preprocessing of the listing data. Columns each function is applied on are required as inputs in the form of lists.
    The following steps are carried out:
    1. Drop columns as specified
    2. Replace text columns with boolean value if the text is filled or not
    3. Replace "t" and "f" with True and False
    4. Split columns where the cells contain lists into multiple columns, one for each unique value in the lists (e.g. amendities) 
    5. Remove currency and percentage from columns to convert them in numerical values
    6. Remove columns without varianve (only one unique value)
    7. Remove columns that only have NAs
    """
    # drop columns as specified
    data = data.drop(columns=cols_to_drop, errors="ignore")
    data = replace_text_with_bool(data, cols_text)
    data = replace_t_f_with_bool(data, cols_t_f)
    data = split_list_into_columns(data, cols_with_list)
    
    data = remove_currency_from_column(data, cols_remove_currency)
    data = remove_percentage_from_column(data, cols_percentage)
    
    # Remove columns without variance
    data = data.loc[:, (data != data.iloc[0]).any()]
    # Remove columns with all NAs
    data = data.dropna(how="all", axis=1)

    return data
```

```python
listings = prepare_listing_data(listings,
                                cols_to_drop,
                                cols_text,
                                cols_t_f,
                                cols_with_list,
                                cols_remove_currency,
                                cols_percentage,)
```

Assuming that the NAs for security deposit and cleaning fee correspond to no fee being charged, the NAs for those columns are replaced with $0.

```python
listings["security_deposit"] = listings["security_deposit"].fillna(0)
listings["cleaning_fee"] = listings["cleaning_fee"].fillna(0)
```

For other columns, a lot of NAs exist, especially for review scores and response rates. As shown below, most rows have no missing values.
Different approaches are feasible in this situation:
1. Remove the rows with a lot of missing values. This will reduce the size of the dataset but since the missing data are focused on only a few rows, they can be removed.
2. Replace the missing values with the median (for numerical values) and the mode (for categorical). The dataset will be skewed but the size remains unchanged

The second approach is carried out for this analysis.

```python
columns_with_na = list(listings.columns[listings.isna().sum() > 0])
```

```python
for col in columns_with_na:
    if listings[col].dtype in ['int', 'float']:
        listings[col].fillna(listings[col].mean(), inplace = True)
    elif listings[col].dtype in ['O', 'bool']:
        listings[col].fillna(listings[col].value_counts().index[0], inplace = True)
```

### Data preparation for the price prediction model
On top of the initial data preparation steps, all categorical columns have to be turned into numerical data. To achieve that. dummies are created for each column that is not numerical and has more than two unique values.

For dates, the values strings are turned into unix epoch timestamps

```python
# Turn dates into timestamps
date_cols = ["host_since", "first_review", "last_review"]
for col in  date_cols:
    listings[col] = pd.to_datetime(listings[col], infer_datetime_format=True)
    listings[col] = listings[col].values.astype(np.int64) // 10 ** 9
```

```python
#turn categorical columns into dummies
cat_cols = list(listings.select_dtypes(include=['object', 'bool']).columns)
for col in  cat_cols:
    listings = pd.concat([listings.drop(col, axis=1), pd.get_dummies(listings[col], prefix=col, prefix_sep='_',
                                                                     drop_first=True)], axis=1)
```

```python
# drop listing id
listings = listings.drop(columns=["id"])
```

### Modelling

```python
#prepare train and test datasets for modelling
test_size = 0.3
rand_state= 42

X = listings.drop(columns=['price'])
y = listings[['price']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=rand_state)
```

```python
#train random forest regressor
rf_model = RandomForestRegressor(n_estimators=100, criterion='mse', random_state=rand_state, n_jobs=-1)
rf_model.fit(X_train, y_train.squeeze())
```

```python
y_test_preds = rf_model.predict(X_test)
y_train_preds = rf_model.predict(X_train)

print('Random Forest MSE train: %.3f, test: %.3f' % (
        mean_squared_error(y_train, y_train_preds),
        mean_squared_error(y_test, y_test_preds)))
print('Random Forest R^2 train: %.3f, test: %.3f' % (
        r2_score(y_train, y_train_preds),
        r2_score(y_test, y_test_preds)))
```

```python
lm_model = LinearRegression(normalize=True)
lm_model.fit(X_train, y_train)

        
y_test_preds = lm_model.predict(X_test)
y_train_preds = lm_model.predict(X_train)

print(r2_score(y_train, y_train_preds))
print(r2_score(y_test, y_test_preds))
print(mean_squared_error(y_test, y_test_preds))
```

```python
#get feature importances from the model
headers = ["name", "score"]
values = sorted(zip(X_train.columns, rf_model.feature_importances_), key=lambda x: x[1] * -1)
forest_feature_importances = pd.DataFrame(values, columns=headers)
forest_feature_importances = forest_feature_importances.sort_values(by = ['score'], ascending = False)

features = forest_feature_importances['name'][:20]
y_pos = np.arange(len(features))
scores = forest_feature_importances['score'][:20]

#plot feature importances
plt.figure(figsize=(10,5))
plt.bar(y_pos, scores, align='center', alpha=0.5)
plt.xticks(y_pos, features, rotation='vertical')
plt.ylabel('Score')
plt.xlabel('Features')
plt.title('Feature importances (Random Forest)')

#plt.savefig('feature importances RF.png')
 
plt.show()
```
